```python
#starting with the boxing script for automatic cropping
from ultralytics import YOLO
from PIL import Image
from os import rename
model = YOLO("yolov8m.pt") #nano version for faster computing

import os
def crop(img_dir):
    results = model.predict(img_dir) #
 
    result = results[0]
    if len(result.boxes) != 1: # remove if zero or multiple boxes
        os.remove(img_dir)
        return
    box = result.boxes[0]
    cords = box.xyxy[0].tolist()
    cords = [round(x) for x in cords]
    class_id = result.names[box.cls[0].item()]

    if str(class_id) != "car" and str(class_id) != "truck":
        print(str(class_id))
        print("Didn't detect car in: " + img_dir)
        print("Removing image ... ")
        
        os.remove(img_dir)
        return
    
    conf = round(box.conf[0].item(), 2)
   
    img= Image.open(img_dir).convert('RGB')
    img = img.crop(cords)
    img.save(img_dir)
    return cords, class_id, conf

current_dir = 'C:/Users/matthew.hui/Documents/code/vans'

# Convertibles: 722 -> 339 (15 min)
# Coupes: 806 -> ~456 (16 min)
# Hatchbacks: 440 hatches 
# Minivans: 994 -> ~536
# Pickups: 600 -> ~322
# Vans: 1048 -> 
# for file in os.listdir(current_dir):
#     crop(os.path.join(current_dir, file))

for i, file in enumerate(os.listdir(current_dir)):
    os.rename(os.path.join(current_dir, file), os.path.join(current_dir, 'van' + str(i) + '.jpg'))


#potential probems: multiple cars and misclassifcation
```
